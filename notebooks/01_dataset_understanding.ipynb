## 9. Target Variable Analysis

The distribution of the target variable is analyzed to determine whether the dataset is balanced. Class imbalance can influence model performance and may require special handling during model training.

```python
df['phishing'].value_counts()
```

```python
df['phishing'].value_counts(normalize=True)*100
```

```python
plt.figure(figsize=(6,4))

sns.countplot(
    data=df,
    x='phishing',
    hue='phishing',
    palette='viridis',
    legend=False
)

plt.title("Distribution of Target Classes")
plt.xlabel("Target Class")
plt.ylabel("Number of Samples")

plt.xticks([0,1], ["Legitimate (0)", "Phishing (1)"])

plt.show()
```

## Observations

- The target variable consists of **two classes**:
  - **0** → Legitimate Website
  - **1** → Phishing Website

- The dataset contains:
  - **5,000 legitimate website samples**
  - **5,000 phishing website samples**

- Both classes contribute **50%** of the total dataset, indicating a **perfectly balanced class distribution**.

- Since the dataset is balanced, there is **no immediate need** for class imbalance handling techniques such as:
  - Random Oversampling
  - Random Undersampling
  - SMOTE (Synthetic Minority Over-sampling Technique)
  - Class Weighting

- A balanced dataset reduces the likelihood of the machine learning model becoming biased toward one class during training.

- This balanced distribution provides a strong foundation for training and evaluating classification models using standard performance metrics.

---

## Key Takeaway

The target variable is **well balanced**, making this dataset suitable for training classification models without requiring additional class balancing techniques during the preprocessing stage.
