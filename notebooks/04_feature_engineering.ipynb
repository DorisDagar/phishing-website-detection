### Objective

The objective of this notebook is to prepare the cleaned phishing website dataset for machine learning by reducing feature redundancy and creating the final engineered dataset.

## 1. Import Libraries

Import the required libraries for feature engineering.

```python
import pandas as pd
import numpy as np
```

## 2. Load the Cleaned Dataset

Load the cleaned dataset generated during the preprocessing stage.

```python
df = pd.read_csv("../dataset/cleaned_phishing_dataset.csv")

df.head()
```

## 3. Separate Features and Target

Separate the input features and the target variable for feature engineering.

```python
X = df.drop("phishing", axis=1)

y = df["phishing"]

print("Features :", X.shape)
print("Target :", y.shape)
```

## 4. Remove Highly Correlated Features

**Question:** Are there any highly correlated features that may provide redundant information?

Identify highly correlated feature pairs and remove one feature from each pair to reduce redundancy in the dataset.

```python
X = df.drop("phishing", axis=1)
y = df["phishing"]
```

```python
# First, calculate the correlation matrix.
corr_matrix = X.corr().abs()

# Extract only the upper triangle to avoid duplicate comparisons.
upper_triangle = corr_matrix.where(
    np.triu(np.ones(corr_matrix.shape), k=1).astype(bool)
)
```

```python
# Find the features whose correlation is greater than 0.80.
features_to_drop = [
    column
    for column in upper_triangle.columns
    if any(upper_triangle[column] > 0.80)
]

X = X.drop(columns=features_to_drop)

print("Removed Features:")
for feature in features_to_drop:
    print("-", feature)

print(f"\nNumber of Features Removed: {len(features_to_drop)}")
print(f"Remaining Features: {X.shape[1]}")
```

### Interpretation

- A correlation threshold of **0.80** was used to identify highly correlated feature pairs.
- One feature from each highly correlated pair was removed to reduce redundancy.
- Although features such as `length_dom` and `length_param` were removed, they were highly correlated with other retained features that capture similar information.
- The final dataset contains **56 features**, making it more compact while preserving the essential information for model building.

---

## 5. Create Final Engineered Dataset

**Question:** What does the final feature set look like after removing redundant features?

Combine the remaining features with the target variable to create the final engineered dataset for model development.


```python
engineered_df = X.copy()

engineered_df["phishing"] = y

engineered_df.head()
```

### Interpretation

- The final engineered dataset contains the remaining informative features after removing highly correlated features.
- This dataset will be used for training and evaluating machine learning models in the next notebook.
